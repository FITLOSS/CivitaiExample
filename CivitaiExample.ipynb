**Load Base Model You Downloaded**

```python
import torch

from diffusers import StableDiffusionPipeline,StableDiffusionXLPipeline
pipeline = StableDiffusionXLPipeline.from_single_file("download_model.safetensors",torch_dtype=torch.float32,use_safetensors=True)
```

**Move model to GPU**

```python
device="cpu"
if torch.cuda.is_available():
    device=torch.device("cuda")
elif torch.backends.mps.is_available():
    device=torch.device("mps")
if device!="cpu":
  pipeline = pipeline.to(device)
else:
  print("You may get error or run long time when you enjoy your model, try switch to GPU machine")
```

**Enjoy Your Model (Base Model)**

```python
pipeline.unload_lora_weights()
torch.manual_seed(0)
image = pipeline(prompt="girl smile with confidence",negative_prompt="ugly").images[0]
image.save("nice_image_1.png")
```

**Enjoy Your Model (Base Model + LoRA)**

```python
pipeline.load_lora_weights("/content/lora_model.safetensors")
torch.manual_seed(0)
image = pipeline(prompt="girl smile with confidence",negative_prompt="ugly").images[0]
image.save("nice_image_1_with_LoRA.png")
```

**Generated Some Images To Your Google Drive**

```python
path_to_your_drive_folder="/content/drive/MyDrive/CivitaiExample"

secret_zip_name="test.zip"
secret_zip_password=b"MyPassword"

LoRA_version_secret_zip_name="LoRA_test.zip"
LoRA_version_secret_zip_password=b"MyPassword"

images_number=10

prompt_without_LoRA="happy girl dress alot"
negative_prompt_without_LoRA="gray , sad"

prompt_with_LoRA="happy girl dress alot"
negative_prompt_with_LoRA="gray , sad"


!pip install pyzipper
import pyzipper
import io,os
```

**No LoRA Images**~

```python
pipeline.unload_lora_weights()
torch.manual_seed(0)


try:
  with pyzipper.AESZipFile(os.path.join(path_to_your_drive_folder,secret_zip_name),'a',compression=pyzipper.ZIP_STORED) as zf:
      zf.setpassword(secret_zip_password)
      zf.setencryption(pyzipper.WZ_AES, nbits=256)
except :
  print("can't create new compress file , maybe it's exist")

for i in range(images_number):
  with pyzipper.AESZipFile(os.path.join(path_to_your_drive_folder,secret_zip_name),'a',compression=pyzipper.ZIP_STORED) as zf:
    zf.setpassword(secret_zip_password)
    zf.setencryption(pyzipper.WZ_AES, nbits=256)
    image = pipeline(prompt=prompt_without_LoRA,negative_prompt=negative_prompt_without_LoRA).images[0]
    file_object = io.BytesIO()
    image.save(file_object, "PNG")
    zf.writestr(str(i)+".png", file_object.getvalue())
```

**With LoRA Images**~

```python
pipeline.load_lora_weights("/content/lora_model.safetensors")
torch.manual_seed(0)


try:
  with pyzipper.AESZipFile(os.path.join(path_to_your_drive_folder,LoRA_version_secret_zip_name),'a',compression=pyzipper.ZIP_STORED) as zf:
      zf.setpassword(LoRA_version_secret_zip_password)
      zf.setencryption(pyzipper.WZ_AES, nbits=256)
except :
  print("can't create new compress file , maybe it's exist")

for i in range(images_number):
  with pyzipper.AESZipFile(os.path.join(path_to_your_drive_folder,LoRA_version_secret_zip_name),'a',compression=pyzipper.ZIP_STORED) as zf:
    zf.setpassword(LoRA_version_secret_zip_password)
    zf.setencryption(pyzipper.WZ_AES, nbits=256)
    image = pipeline(prompt=prompt_with_LoRA,negative_prompt=negative_prompt_with_LoRA).images[0]
    file_object = io.BytesIO()
    image.save(file_object, "PNG")
    zf.writestr(str(i)+".png", file_object.getvalue())
```
